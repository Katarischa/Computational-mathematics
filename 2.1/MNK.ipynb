# Задача 1

Дана выборка $y_i$. Требуется решить задачу МНК минимизируя функцию потерь 
$$\mathfrak{L} = \sum\limits_{i  =1}^l(y_i - \check{y})^2 \longrightarrow min_{\check{y}}$$

### Решение:

Т. к. $\check{y} = const$, то Из минимизации фугкции потерь имеем:
$$\frac{\partial \mathfrak{L}}{\partial \check{y}} = \sum\limits_{i  =1}^l2(y_i - \check{y}) = 0 \Longrightarrow \check{y} = \frac{1}{l}\sum\limits_{i  =1}^ly_i$$

# Задача 3

Покажите, что прямая, пострроенная по методу МНК, всегда проходит через точку $(\overline{x},\overline{y})$. Обобщить на многомерный случай.

### Решение:

1. Для одномерной регресии утверждение очевидно, поскольку для модели $\check{y} = w_1 x+w_0$, свободный член по определению равен $\hat{w_0}:=\overline{y} - \hat{w} x$
2. В многомерном случае для нескольких $y$ утверждение так же очевидно выполнено из предыдущего шага, т. к. каждый из $y$ можно рассмотреть отдельно, т е свести к одномерному случаю.

Для нескольких $x$ имеем следующее: $y_i = \hat{w_0}  + \hat{w_1}x_{i1} + ... + \hat{w_k}x_{ik}$. Сложим все измерения и поедлим на $l$:
$$\overline{y} = \frac{1}{l}\sum\limits_1^ly_i = \hat{w_0}  + \frac{1}{l}\hat{w_1}\sum\limits_1^lx_{i1} + ... + \frac{1}{l}\hat{w_k}\sum\limits_1^lx_{ik} = \hat{w_0}  + \hat{w_1}\overline{x_{1}} + ... + \hat{w_k}\overline{x_{k}}$$
Т е полученная из МНК гиперплоскость проходит через точку $(\overline{x}_1,..,\overline{x}_k,\overline{y})$

### Решение:

По условию $Xw = y$ - недоопределенная система, при этом $X$ имеет максимальный ранг. Хотим минимизировать $\|w\|_2  =w^Tw$. Рассмотрим следующую функцию потерь:
$$\mathfrak L = \frac{1}{2}w^Tw+a(Xw-y)$$
где $a$ - некоторая строка, такая что $a(Xw-y)\in \mathfrak{R}$. А коэффициент $1/2$ чисто для красоты ответа. Дифференцируем:
$$\frac{\partial\mathfrak L}{\partial w} = w + X^Ta^T = 0 \Longrightarrow \begin{cases}w = -X^Ta^T\\y=Xw \end{cases} \Longrightarrow y = -XX^Ta^T \Longrightarrow a^T = -(XX^T)^{-1}y$$
Итого $w  =X^T(XX^T)^{-1}y$. Видно, что при $y$ стоит правая псевдообратная матрица.

# Задача 8

$$\widetilde{y} \sim P(\widetilde{y}) = \frac{1}{(2\pi)^{l/2}\text{det}A}\exp\left(-\frac{(\widetilde{y}-y)^TA(\widetilde{y}-y)}{2}\right)$$

1. Сделаем замену $\widetilde{y}-y = Sz$, где $S^TAS = E \Longrightarrow \text{det}S = \sqrt{\text{det}A}$. Ипсользуя эти выкладки проинтегрируем функцию распределения:
$$\int P(\widetilde{y})d^l\widetilde{y} = \int \frac{1}{(2\pi)^{l/2}}\exp\left(-\frac{z^Tz}{2}\right)d^lz = \left[\int\frac{1}{\sqrt{2\pi}}\exp\left(-\frac{1}{2}u^2_1\right)du_1\right]...\left[\int\frac{1}{\sqrt{2\pi}}\exp\left(-\frac{1}{2}u^2_l\right)du_l\right] = |\text{интеграл Эйлера}| = 1^l = 1$$

2. На курсе теорвера было доказано, что определение гауссовского вектора эквивалентно утверждению: $\forall t\in\mathfrak{R}^n\hookrightarrow \langle t,\widetilde{y}\rangle$ - нормально распределенная случ величина ($\sim\mathcal{N}(a_t,\sigma_t^2)$). Найдем параметры этого распределения:
$$a_t = E\langle t,\widetilde{y}\rangle  =E\sum\limits_1^lt_i\widetilde{y}_i = \sum\limits_1^l t_iE\widetilde{y}_i = \langle t,E\widetilde{y}\rangle$$
$$\sigma_t^2 = D\langle t, \widetilde{y}\rangle = D\sum\limits_1^l t_i\widetilde{y}_i = cov\left(\sum t_i\widetilde{y}_i,\sum t_j\widetilde{y}_j \right) = \sum_{i,j=1}^lcov(\widetilde{y}_i,\widetilde{y}_j)t_it_j = \langle D\widetilde{y}t,t\rangle,\;\;\;\;D\widetilde{y} = \|cov(\widetilde{y}_i,\widetilde{y}_j)\|_{i,j = 1}^l$$
Тогда характеристическая функция величины $\widetilde{y}$ в точке $t$ принимает значение:
$$P_{\widetilde{y}}(t) \sim Ee^{i\langle t,\widetilde{y}\rangle} = P_{\langle t,\widetilde{y}\rangle}(1) = \exp\left(ia_t - \frac{1}{2}\sigma_t^2\right) = \exp\left(i\langle t,E\widetilde{y}\rangle-\frac{1}{2}\langle D\widetilde{y}t,t\rangle\right)$$
Сопоставляя с данным в условии распределением, получаем, что $A = D\widetilde{y}$ - матрица ковариаций $\Longrightarrow cov(\widetilde{y}_i,\widetilde{y}_j) = A_{ij}$

3.  Обозначим псевдообратную левую матрицу за $Q = (X^TX)^{-1}X^T$. Тогда $\widetilde{w}_{\alpha} = Q^{\alpha i}\widetilde{y}_i$. Согласно лекции:
$$\sigma_{w_{\alpha}}^2 = <<\widetilde{w}_{\alpha},\widetilde{w}_{\alpha}>>= Q^{\alpha i}Q^{j \alpha }<<\widetilde{y}_i,\widetilde{y}_j>>= Q^{\alpha i}Q^{\alpha j}cov(\widetilde{y}_i,\widetilde{y}_j) = Q^{\alpha i}Q^{\alpha j}A_{ij} = \left(QAQ^T\right)_{\alpha\alpha}$$

4. $A = \text{diag}(A_1,A_2,..,A_l) \Longrightarrow \sigma^2_{w_{\alpha}} = Q^{\alpha i}Q^{i\alpha }A_{ii}$ Осталось найти матрицу $Q$:
$$X = \begin{bmatrix}x_1 & 1\\..&..\\x_l&1\end{bmatrix}\Longrightarrow X^TX =\begin{bmatrix}x_1 & .. &x_l\\1&..&1\end{bmatrix} \begin{bmatrix}x_1 & 1\\..&..\\x_l&1\end{bmatrix}=l\begin{bmatrix}\overline{x^2} & \overline{x}\\\overline{x}&1\end{bmatrix}\Longrightarrow (X^TX)^{-1} = \frac{1}{l\:var x}\begin{bmatrix}1 & -\overline{x}\\-\overline{x}&\overline{x^2}\end{bmatrix}$$
$$Q = \frac{1}{l\:var x}\begin{bmatrix}1 & -\overline{x}\\-\overline{x}&\overline{x^2}\end{bmatrix}\begin{bmatrix}x_1 & .. &x_l\\1&..&1\end{bmatrix} = \frac{1}{l\:var x}\begin{bmatrix} x_1-\overline{x}&..&x_l-\overline{x}\\ \overline{x^2}-x_1&..&\overline{x^2}-x_l\end{bmatrix}$$
Тогда:
$$\sigma^2_{w_{1}} = \frac{1}{l^2\:var^2 x}\sum\limits_{i=0}^lA_i\left(x_i - \overline{x}\right)^2$$
$$\sigma^2_{w_{2}} = \frac{1}{l^2\:var^2 x}\sum\limits_{i=0}^lA_i\left( \overline{x^2} - x_i\overline{x}\right)^2$$

Что касается y-погрешности для i-ой точки ($A_i = s_i^2$) :
$$var\:\widetilde{y}_i = cov(\widetilde{y}_i,\widetilde{y}_i) = s_i^2 \Longrightarrow \sigma_{\widetilde{y}_i} = s_i$$
